```python
import pandas as pd
from newspaper import Article, Config
from GoogleNews import GoogleNews

user_agent = 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'
config = Config()
config.browser_user_agent = user_agent

nltk.download('punkt')
googlenews = GoogleNews(start='04/01/2023', end='04/05/2023')
googlenews.search('crypto currency')
result = googlenews.result()
df = pd.DataFrame(result)
print(df.head())

articles_list = []
for i in range(2, 10):
    googlenews.getpage(i)
    result = googlenews.result()
    df = pd.DataFrame(result)
    for ind in df.index:
        dict = {}
        try:
            article = Article(df['link'][ind], config=config)
            article.download()
            article.parse()
            article.nlp()
            dict['Date'] = df['date'][ind]
            dict['Media'] = df['media'][ind]
            dict['Title'] = article.title
            dict['Article'] = article.text
            dict['Summary'] = article.summary
            articles_list.append(dict)
        except Exception as e:
            print(e)

news_df = pd.DataFrame(articles_list)
news_df.to_excel("articles.xlsx")
```

```python
from GoogleNews import GoogleNews
from newspaper import Article
from newspaper import Config
import pandas as pd

# Set up user agent and Google News API
user_agent = 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'
config = Config()
config.browser_user_agent = user_agent
googlenews = GoogleNews()

# Set start and end dates
start_date = '2023-05-01'
end_date = '2023-05-05'

# Search for articles on a specific topic between the start and end dates
googlenews.search('crypto currency')
googlenews.set_time_range(start_date, end_date)
result = googlenews.result()
df = pd.DataFrame(result)

# Loop through all pages of search results and extract article content
articles = []
for i in range(2, 10):
    googlenews.getpage(i)
    result = googlenews.result()
    df = pd.DataFrame(result)
    for ind in df.index:
        dict = {}
        article = Article(df['link'][ind], config=config)
        try:
            article.download()
            article.parse()
            article.nlp()
        except:
            continue
        dict['Date'] = df['date'][ind]
        dict['Media'] = df['media'][ind]
        dict['Title'] = article.title
        dict['Article'] = article.text
        dict['Summary'] = article.summary
        articles.append(dict)

# Create pandas dataframe and save as Excel file
news_df = pd.DataFrame(articles)
news_df.to_excel("articles.xlsx")
```
